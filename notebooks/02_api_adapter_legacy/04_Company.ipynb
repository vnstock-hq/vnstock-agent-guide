Demo tự động các hàm trong class `Company`.

```python
from vnstock_data.api.company import Company

# Khởi tạo instance
instance = Company(source='VCI', symbol='VNM')
```

### `affiliate()`
Retrieve company affiliate data.

```python
df_affiliate = instance.affiliate()
display(df_affiliate)
```

### `capital_history()`
Retrieve company charter capital history.

```python
df_capital_history = instance.capital_history()
display(df_capital_history)
```

### `events()`
Retrieve company events.

```python
df_events = instance.events()
display(df_events)
```

### `news()`
Retrieve company news.

```python
df_news = instance.news()
display(df_news)
```

### `officers()`
Retrieve company officers data.

```python
df_officers = instance.officers()
display(df_officers)
```

### `overview()`
Retrieve company overview data.

```python
df_overview = instance.overview()
display(df_overview)
```

### `shareholders()`
Retrieve company shareholders data.

```python
df_shareholders = instance.shareholders()
display(df_shareholders)
```

### `subsidiaries()`
Retrieve company subsidiaries data.

```python
df_subsidiaries = instance.subsidiaries()
display(df_subsidiaries)
```
