Demo tự động các hàm trong class `TopStock`.

```python
from vnstock_data.api.insight import TopStock

# Khởi tạo instance
instance = TopStock(source='VND', symbol='VNM')
```

### `deal()`
Top 10 by block trade volume in the given index.

```python
df_deal = instance.deal()
display(df_deal)
```

### `foreign_buy()`
Top 10 net foreign buys on the given date.

```python
df_foreign_buy = instance.foreign_buy()
display(df_foreign_buy)
```

### `foreign_sell()`
Top 10 net foreign sells on the given date.

```python
df_foreign_sell = instance.foreign_sell()
display(df_foreign_sell)
```

### `gainer()`
Top 10 gainers in the given index.

```python
df_gainer = instance.gainer()
display(df_gainer)
```

### `loser()`
Top 10 losers in the given index.

```python
df_loser = instance.loser()
display(df_loser)
```

### `value()`
Top 10 by trading value in the given index.

```python
df_value = instance.value()
display(df_value)
```

### `volume()`
Top 10 by abnormal volume in the given index.

```python
df_volume = instance.volume()
display(df_volume)
```
