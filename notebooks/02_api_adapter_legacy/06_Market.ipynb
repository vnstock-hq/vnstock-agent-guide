Demo tự động các hàm trong class `Market`.

```python
from vnstock_data.api.market import Market

# Khởi tạo instance
instance = Market(source='VND', symbol='VNM')
```

### `evaluation()`
Retrieves combined P/E and P/B data for the given duration.

```python
df_evaluation = instance.evaluation()
display(df_evaluation)
```

### `pb()`
Retrieves P/B ratio data for the given duration.

```python
df_pb = instance.pb()
display(df_pb)
```
