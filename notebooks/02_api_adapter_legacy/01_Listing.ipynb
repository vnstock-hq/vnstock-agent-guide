Demo tự động các hàm trong class `Listing`.

```python
from vnstock_data.api.listing import Listing

# Khởi tạo instance
instance = Listing(source='VCI')
```

### `all_bonds()`
Retrieve all bonds (group='BOND').

```python
df_all_bonds = instance.all_bonds()
display(df_all_bonds)
```

### `all_covered_warrant()`
Retrieve all covered warrants (group='CW').

```python
df_all_covered_warrant = instance.all_covered_warrant()
display(df_all_covered_warrant)
```

### `all_etf()`
Retrieve all ETF (exchange-traded funds).

```python
df_all_etf = instance.all_etf()
display(df_all_etf)
```

### `all_future_indices()`
Retrieve all futures indices (group='FU_INDEX').

```python
df_all_future_indices = instance.all_future_indices()
display(df_all_future_indices)
```

### `all_government_bonds()`
Retrieve all government bonds.

```python
df_all_government_bonds = instance.all_government_bonds()
display(df_all_government_bonds)
```

### `all_indices()`
Retrieve all standardized market indices with metadata.

```python
df_all_indices = instance.all_indices()
display(df_all_indices)
```

### `all_symbols()`
Retrieve all symbols (filtered to STOCK).

```python
df_all_symbols = instance.all_symbols()
display(df_all_symbols)
```

### `indices_by_group()`
Retrieve standardized market indices by group/category.

```python
df_indices_by_group = instance.indices_by_group()
display(df_indices_by_group)
```

### `industries_icb()`
Retrieve ICB code hierarchy and mapping.

```python
df_industries_icb = instance.industries_icb()
display(df_industries_icb)
```

### `symbols_by_exchange()`
Retrieve symbols by exchange/board.

```python
df_symbols_by_exchange = instance.symbols_by_exchange()
display(df_symbols_by_exchange)
```

### `symbols_by_group()`
Retrieve symbols by predefined group (VN30, HNX30, CW, etc.).

```python
df_symbols_by_group = instance.symbols_by_group()
display(df_symbols_by_group)
```

### `symbols_by_industries()`
Retrieve symbols grouped by ICB industries.

```python
df_symbols_by_industries = instance.symbols_by_industries()
display(df_symbols_by_industries)
```
