Demo tự động các hàm trong class `Finance`.

```python
from vnstock_data.api.financial import Finance

# Khởi tạo instance
instance = Finance(source='vci', symbol='VNM')
```

### `balance_sheet()`
Retrieve balance sheet data.

```python
df_balance_sheet = instance.balance_sheet(period='quarter')
display(df_balance_sheet)
```

### `cash_flow()`
Retrieve cash flow statement data.

```python
df_cash_flow = instance.cash_flow(period='quarter')
display(df_cash_flow)
```

### `income_statement()`
Retrieve income statement data.

```python
df_income_statement = instance.income_statement(period='quarter')
display(df_income_statement)
```

### `ratio()`
Retrieve financial ratios.

```python
df_ratio = instance.ratio(period='quarter')
display(df_ratio)
```

### `note()`
Retrieve financial statement notes (thuyết minh báo cáo tài chính) if source is 'vci'.

```python
df_note = instance.note(period='quarter')
display(df_note)
```
