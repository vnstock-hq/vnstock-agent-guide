Demo module `vnstock_data.api.trading` với các API lấy số liệu giao dịch, lịch sử giá và dữ liệu Market.

```python
from vnstock_data.api.trading import Trading
```

## 1. Dữ liệu Giao Dịch Nhà Đầu Tư Nước Ngoài (Foreign Trade)
**Hỗ trợ bởi Nguồn:** VCI, CAFEF
Mô tả: Lấy giao dịch khối ngoại theo lịch sử.

```python
# Sử dụng nguồn VCI
vci_trading = Trading(source='VCI', symbol='VCI')
df_foreign_trade = vci_trading.foreign_trade(start='2024-01-01', end='2024-01-10')
display(df_foreign_trade.head())
```

## 2. Giao Dịch Tự Doanh (Proprietary Trade)
**Hỗ trợ bởi Nguồn:** VCI, CAFEF
Mô tả: Lấy dữ liệu tự doanh công ty chứng khoán.

```python
df_prop_trade = vci_trading.prop_trade()
display(df_prop_trade.head())
```

## 3. Giao Dịch Nội Bộ (Insider Deal)
**Hỗ trợ bởi Nguồn:** VCI, CAFEF
Mô tả: Lịch sử giao dịch nội bộ và cổ đông lớn.

```python
df_insider_deal = vci_trading.insider_deal(limit=10)
display(df_insider_deal.head())
```

## 4. Lịch Sử Giá (Price History)
**Hỗ trợ bởi Nguồn:** VCI, KBS, CAFEF
Mô tả: Lấy lịch sử giá (OHLCV) hàng ngày/phút/tuần/tháng tuỳ nguồn hỗ trợ.

```python
# Chuyển sang dùng KBS
kbs_trading = Trading(source='KBS', symbol='VCI')
df_price_history = kbs_trading.price_history(start='2024-01-01', end='2024-01-15')
display(df_price_history.head())
```

## 5. Bảng Giá Đang Khớp & Cổ Phiếu Dễ Mua Dễ Bán (Price Board / Odd Lot / Put Through)
**Hỗ trợ bởi Nguồn:** VCI, KBS
Mô tả: Xem bảng giá snapshot realtime, các lệnh lô lẻ (odd_lot) hoặc thoả thuận (put_through).

```python
df_price_board = vci_trading.price_board(symbols_list=['SSI', 'VCI', 'HCM'])
display(df_price_board.head())

# Giao dịch thoả thuận
df_put_through = vci_trading.put_through(exchange='HOSE')
display(df_put_through.head())
```

## 6. Lịch Sử Kê Lệnh Bổ Sung (Trade history / matched_by_price)
**Hỗ trợ bởi Nguồn:** KBS
Mô tả: Phân rã lệnh trong một phiên, hoặc lệnh khớp theo các mức giá. 

```python
# Phân bổ lệnh theo giá trong phiên (matched_by_price) của mã VCI
df_matched = kbs_trading.matched_by_price()
display(df_matched.head())

# Chi tiết từng lệnh khớp (trade_history)
df_trade = kbs_trading.trade_history(limit=5)
display(df_trade)
```
