Demo tự động các hàm trong class `CommodityPrice`.

```python
from vnstock_data.api.commodity import CommodityPrice

# Khởi tạo instance
instance = CommodityPrice(source='SPL')
```

```python
df_coke = instance.coke()
display(df_coke)
```
