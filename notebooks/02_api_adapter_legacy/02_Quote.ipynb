Demo tự động các hàm trong class `Quote`.

```python
from vnstock_data.api.quote import Quote

# Khởi tạo instance
instance = Quote(source='KBS', symbol='VNM')
```

### `history()`
Load historical OHLC data for the symbol.

```python
df_history = instance.history(length=30)
display(df_history)
```

### `intraday()`
Load intraday trade data for the symbol.

```python
df_intraday = instance.intraday()
display(df_intraday)
```

### `price_depth()`
Load price depth (order book) data for the symbol.

```python
df_price_depth = instance.price_depth()
display(df_price_depth)
```
