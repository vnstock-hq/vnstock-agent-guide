````python
from vnstock_data import Insights
from IPython.display import display

# Initialize Insights layer
ins = Insights()
````

### Cấu trúc Khai Thác Hiện Tại

Insights API được quy hoạch vào 2 nhánh tác vụ (Domains):
- `ranking()`: Xếp hạng top biến động dòng tiền / giá.
- `screener()`: Bộ lọc truy vấn chứng khoán theo hàng chục tiêu chí.

### 1. Ranking Domain (Xếp Hạng & Biến Động Top Cổ Phiếu)

Nhận diện dấu chân của dòng tiền thông minh trên các báo cáo biến động tăng/giảm, khối lượng đột biến và hoạt động của Khối ngoại.

**Top cổ phiếu tăng giá mạnh nhất (Top Gainers - VNINDEX)**

````python
df = ins.ranking().gainer(index="VNINDEX", limit=5)
display(df)
````

**Top cổ phiếu giảm giá mạnh nhất toàn thị trường (Top Losers)**

````python
df = ins.ranking().loser(limit=5)
display(df)
````

**Top cổ phiếu thanh khoản cao nhất theo Giá trị (Top by Value)**

````python
df = ins.ranking().value(limit=5)
display(df)
````

**Top cổ phiếu biến động khối lượng cực đại (Top by Volume Spike)**

````python
df = ins.ranking().volume(limit=5)
display(df)
````

**Top cổ phiếu được Nước ngoài Mua Ròng (Top Foreign Net Buy)**

````python
import datetime
today_str = datetime.datetime.now().strftime('%Y-%m-%d')
df_foreign_buy = ins.ranking().foreign_buy(date=today_str, limit=5)
display(df_foreign_buy)
````

**Top cổ phiếu bị Nước ngoài Bán Ròng (Top Foreign Net Sell)**

````python
df_foreign_sell = ins.ranking().foreign_sell(date=today_str, limit=5)
display(df_foreign_sell)
````

**Top hoạt động mua bán Giao dịch thỏa thuận (Top Deals Put-through)**

````python
df_deals = ins.ranking().deal(limit=5)
display(df_deals)
````

### 2. Screener Domain (Bộ Lọc Quét Cổ Phiếu Dataframe)

Kéo thẳng toàn bộ Snapshot thị trường (>1500 mã) với hàng chục tiêu chí cơ bản/kỹ thuật. Nhà đầu tư không cần học cú pháp SQL phức tạp, mà có thể dùng sức mạnh của core thư viện `pandas` để filter tuỳ chỉnh.

**Tra cứu từ điển các định nghĩa biến số bộ lọc (Dictionary of Filter Criteria)**

````python
df_criteria = ins.screener().criteria(lang="vi")
display(df_criteria.head(10))
````

**Tải bảng dữ liệu toàn bộ cổ phiếu trên 3 sàn (Full Market Snapshots - 2000 tickers)**

````python
df_all = ins.screener().filter(limit=2000)
display(df_all.head())
````
