````python
from vnstock_data import Macro
from IPython.display import display

# Initialize Macro layer
mac = Macro()
````

### Cấu Trúc Khai Thác Mới (từ Vnstock 3.0.0)

Macro API hiện được tổ chức gọn gàng thành 3 nhóm (Domains) chính:

- `economy()`
- `currency()`
- `commodity()`

*(Lưu ý: Các hàm gọi trực tiếp phân cấp cũ như `mac.gdp()` sẽ bị loại bỏ vào T8/2026).*.

### 1. Economy Domain (Chỉ Báo Kinh Tế Việt Nam)

Các chỉ báo định lượng hiệu suất của nền kinh tế: GDP, CPI, FDI, Sản xuất công nghiệp, Xuất Nhập Khẩu, Doanh thu bán lẻ...

**Tăng trưởng GDP hàng quý (GDP Quarterly Growth)**

````python
df = mac.economy().gdp(period="quarter", length=4)
display(df)
````

**Chỉ số giá tiêu dùng CPI (Consumer Price Index)**

````python
df = mac.economy().cpi(period="month", length=3)
display(df)
````

**Đầu tư trực tiếp nước ngoài FDI (Foreign Direct Investment)**

````python
df = mac.economy().fdi(period="month", length=3)
display(df)
````

**Tình hình Xuất Nhập khẩu và Cán cân thương mại (Import/Export & Trade Balance)**

````python
df = mac.economy().import_export(period="month", length=3)
display(df)
````

### 2. Currency Domain (Tiền Tệ & Lãi Suất)

Theo dõi Tỷ giá hối đoái (FX) và Lãi suất (Interest Rates) liên ngân hàng / huy động.

**Tỷ giá hối đoái trung tâm (Exchange Rates - USD, EUR, JPY)**

````python
df = mac.currency().exchange_rate(period="day", length=5)
display(df)
````

**Lãi suất ngân hàng (Interest Rates - Lending & Deposit)**

````python
df = mac.currency().interest_rate(period="day", length=90)
display(df)
````

### 3. Commodity Domain (Thị Trường Hàng Hoá)

Truy xuất giá Dầu thô, Vàng, Thép, Khí tự nhiên, Nông sản... trên thị trường nội địa (VN) và thế giới (GLOBAL).

**So sánh giá Vàng SJC (VN) và Vàng Thế Giới (Global Gold Futures)**

````python
df_vn = mac.commodity().gold(market="VN", length=5)
df_global = mac.commodity().gold(market="GLOBAL", length=5)
display(df_vn)
display(df_global)
````

**Giá dầu thô WTI và Brent (Crude Oil)**

````python
df = mac.commodity().oil_crude(length=5)
display(df)
````

**Giá thép nội địa (Thép D10) và Thép HRC thế giới (Steel Markets)**

````python
df_vn = mac.commodity().steel(market="VN", length=5)
df_global = mac.commodity().steel(market="GLOBAL", length=5)
display(df_vn)
display(df_global)
````

**Giá thịt heo hơi nội địa (Vietnam Live Pork Price)**

````python
df = mac.commodity().pork(market="VN", length=5)
display(df)
````
