````python
from vnstock_data import Fundamental
from IPython.display import display

# Initialize Fundamental layer
fun = Fundamental()
````

### Cấu trúc Gọi Hàm (Calling Conventions)

Mô đun Fundamental hỗ trợ 2 dạng cú pháp linh hoạt (Proxy chaining):
1. `fun.equity("MÃ").method()` - Hướng đối tượng đối với một mã.
2. `fun.equity.method("MÃ")` - Dạng hàm tiện ích (Utility function).

### 1. Income Statement (Báo Cáo Kết Quả Kinh Doanh)

Báo cáo thu nhập, doanh thu, chi phí, lợi nhuận hoạt động và lợi nhuận ròng qua các kỳ.

**Lấy báo cáo kết quả KD ngân hàng TCB (Income Statement - TCB)**

````python
df = fun.equity("TCB").income_statement(limit=4)
display(df.head())
````

### 2. Balance Sheet (Bảng Cân Đối Kế Toán)

Số liệu phản ánh tài sản, nợ phải trả, vốn chủ sở hữu qua các kỳ báo cáo.

**Lấy cân đối kế toán doanh nghiệp VIC (Balance Sheet - VIC)**

````python
df = fun.equity.balance_sheet("VIC", limit=4)
display(df.head())
````

### 3. Cash Flow (Báo Cáo Lưu Chuyển Tiền Tệ)

Theo dõi dòng tiền từ hoạt động kinh doanh, đầu tư và tài chính.

**Lưu chuyển tiền tệ của VNM (Cash Flow - VNM)**

````python
df = fun.equity("VNM").cash_flow(limit=4)
display(df.head())
````

### 4. Financial Ratio (Tỷ Số Tài Chính)

Tổng hợp các tỷ số quan trọng như P/E, P/B, ROA, ROE, Tỷ lệ vay nợ (Debt/Equity), v.v.

**Tỷ số tài chính của HPG (Financial Ratios - HPG)**

````python
df = fun.equity("HPG").ratio(limit=4)
display(df.head())
````

### 5. Footnotes (Thuyết Minh BCTC)

Các văn bản thuyết minh, ghi chú báo cáo tài chính đính kèm để làm rõ các số liệu.

**Trích xuất Thuyết minh BCTC của FPT (Financial Disclosures/Notes - FPT)**

````python
df = fun.equity("FPT").note()
display(df.head())
````

### 6. Financial Health / Unified Scorecard (Sức Khoẻ Tổng Hợp BCTC)

Tính năng cao cấp nhất (*từ phiên bản Vnstock 3.1.0*). Chức năng này tự động đánh giá mã ngành ICB cấp 2 của công ty, **ép** dữ liệu từ 3 báo cáo (KQKD, CĐKT, LCTT) và Tỷ số tài chính chay qua một bộ lọc chuẩn (Scorecard). Kết quả sẽ cho ra một bảng nối ngang với format và thuật ngữ tiếng Việt cực chuẩn (lấy cảm hứng từ TCBS).

**Báo cáo hợp nhất tự động phát hiện mẫu Ngành Ngân Hàng (Auto Scorecard - Bank - VI)**

````python
df = fun.equity("TCB").financial_health(scorecard="auto", lang="vi", limit=4)
display(df.head().T)
````

**Báo cáo hợp nhất tự động phát hiện mẫu Ngành Chứng Khoán (Auto Scorecard - Securities - VI)**

````python
df = fun.equity("SSI").financial_health(scorecard="auto", lang="vi", limit=4)
display(df.head().T)
````

````python
df = fun.equity("BVH").financial_health(scorecard="auto", lang="vi", limit=4)
display(df.head().T)
````
