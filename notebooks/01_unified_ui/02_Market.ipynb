## 1. Equity Market / Thị Trường Cổ Phiếu

Retrieve OHLCV, trade history, order book, and other statistics for individual stocks.

*Lấy giá lịch sử, lịch sử lệnh, sổ lệnh, và các thống kê khác cho từng cổ phiếu.*

### Historical Price (OHLCV) / Lịch sử giá

````python
df_ohlcv = mkt.equity("VIC").ohlcv(start="2026-02-01", end="2026-03-01")
display(df_ohlcv.head())
````

### Trade History / Lịch sử Khớp Lệnh
Daily match and deal statistics. / *Thống kê khớp lệnh và thỏa thuận theo ngày.*

````python
history_stats = mkt.equity("VIC").trade_history(start="2026-02-01", end="2026-03-01")
display(history_stats.head())
````

### Intraday Trades / Bảng Chi tiết Lệnh (Time & Sales)

````python
df_trades = mkt.equity("TCB").trades()
display(df_trades.head())
````

### Order Book / Cấp độ Mua/Bán
Real-time top bid/ask levels. / *Mức giá chào mua/bán tốt nhất theo thời gian thực.*

````python
df_orderbook = mkt.equity("VNM").order_book()
display(df_orderbook.head())
````

### Quote / Bảng giá

````python
quote = mkt.equity("HPG").quote()
display(quote)
````

### Session Statistics / Thống kê Phiên
Intraday session volume and value metrics. / *Chỉ số khối lượng và giá trị trong phiên giao dịch.*

````python
session = mkt.equity("VIC").session_stats()
display(session)
````

### Equity Summary / Tổng hợp Cổ phiếu

````python
summary_info = mkt.equity("VIC").summary()
display(summary_info)
````

### Market Flow & Intraday Details / Dòng tiền và Thông tin chi tiết
Foreign flow, Proprietary trading flow, Block trades, Odd lots, and Volume Profile. / *Dòng tiền ngoại, dòng tiền tự doanh, giao dịch thỏa thuận, giao dịch lô lẻ và phân bố khối lượng theo giá.*

````python
# Foreign flow / Dòng tiền ngoại
foreign = mkt.equity("VIC").foreign_flow()
display(foreign.head() if not foreign.empty else "No data")

# Proprietary flow / Dòng tiền tự doanh
proprietary = mkt.equity("VIC").proprietary_flow()
display(proprietary.head() if not proprietary.empty else "No data")

# Block trades / Giao dịch thỏa thuận
blocks = mkt.equity("VIC").block_trades()
display(blocks.head() if not blocks.empty else "No data")

# Odd lot trades / Giao dịch lô lẻ
odds = mkt.equity("VIC").odd_lot()
display(odds.head() if not odds.empty else "No data")

# Volume profile / Phân bố khối lượng theo giá
vol_profile = mkt.equity("VJC").volume_profile()
display(vol_profile.head() if not vol_profile.empty else "No data")
````

## 3. Crypto

Sử dụng bộ Unified UI để lấy dữ liệu Spot từ thị trường Crypto (nguồn: sàn **Binance**). Hỗ trợ các endpoints như OHLCV, Sổ lệnh (Order Book), Lịch sử giao dịch (Intraday), và Báo giá nhanh (Quote).

````python
from typing import Any
from vnstock_data import Market
from IPython.display import display

# Khởi tạo đối tượng Market và lấy dữ liệu cặp tiền số BTCUSDT qua domain Crypto
mkt = Market()
crypto = mkt.crypto('BTCUSDT')

print("-- 1. Ticker Quote --")
display(crypto.quote())
````

````python
print("-- 2. OHLCV Lịch sử (5 Nến ngày gần nhất) --")
display(crypto.ohlcv(interval='1d', limit=5))
````

````python
print("-- 3. Sổ Lệnh (Order Book - L10) --")
# Lấy 5 cấp độ sổ lệnh gần nhất theo chuẩn schema thiết kế
display(crypto.order_book(limit=5))
````

````python
print("-- 4. Lịch sử giao dịch Intraday --")
display(crypto.intraday(limit=5))
````

### Aggregate Trades – Lệnh gộp
Truyền `mode='aggregate'` vào `trades()` để lấy dữ liệu gộp (các fills cùng giá và chiều được nén lại).

*Binance: `GET /api/v3/aggTrades`*

````python
df_agg = crypto.trades(limit=5, mode='aggregate')
display(df_agg)
````

### Các phương thức mở rộng (Extended Methods)

Domain **Crypto** cung cấp thêm các phương thức theo chuẩn quốc tế, với tên gọi chuẩn FIX/Bloomberg:

| Method | Mô tả | Binance Endpoint |
|---|---|---|
| `ohlcv_raw` | Nến thô đầy đủ fields (`quote_volume`, `trades`...) | `GET /api/v3/klines` |
| `trade_history` | Lịch sử lệnh theo ID, phân trang | `GET /api/v3/historicalTrades` |
| `vwap` | Giá trung bình VWAP | `GET /api/v3/avgPrice` |
| `daily_stats` | Thống kê phiên trong ngày | `GET /api/v3/ticker/tradingDay` |
| `last_price` | Giá khớp gần nhất (LTP) – cực nhẹ | `GET /api/v3/ticker/price` |
| `rolling_stats` | Thống kê biến động theo cửa sổ thời gian | `GET /api/v3/ticker` |
| `reference_price` | Giá tham chiếu (mark/liquidation) | `GET /api/v3/referencePrice` |

````python
# OHLCV Raw – Nến thô với quote_volume, trade count, taker buy fields
df_raw = crypto.ohlcv(mode='raw', interval='1h', limit=3)
display(df_raw)
````

````python
# VWAP – Giá trung bình có trọng số khối lượng
display(crypto.vwap())
````

````python
# Last Price – Payload cực nhẹ, chỉ lấy giá hiện tại
print(crypto.last_price())
````

````python
# Rolling Stats – Thống kê biến động giá trượt (ví dụ cửa sổ 1 giờ)
display(crypto.rolling_stats(window_size='1h').head())
````

````python
# Daily Stats – Thống kê toàn phiên trong ngày
display(crypto.daily_stats())
````

## 4. Khám phá thị trường Ngoại hối & Hàng hoá Quốc tế

Sử dụng nguồn dữ liệu **Dukascopy** để lấy dữ liệu lịch sử nén delta tốc độ cao cho ngoại tệ, kim loại quý và chỉ số.
Các tham số:
- `interval`: Hỗ trợ chuẩn `1m`, `1h`, `4h`, `1d`, `1w`, `1M`. Có cơ chế gom nến tự động.
- `side`: Tùy chọn lấy giá `bid` (mặc định), `ask` hoặc `mid`.


````python
print("-- 1. Tỷ giá Ngoại Hối EUR/USD (Interval 1 giờ, 5 nến) --")
df_forex = mkt.forex("EURUSD").ohlcv(interval="1h", length=5, side="mid")
display(df_forex.head())

print("-- 2. Lịch sử giá Vàng XAU/USD (Interval 1d, 5 nến) --")
# Lưu ý: Mặc định ForexMarket gọi tới nhóm FX_METAL cho kim loại quý
df_gold = mkt.forex("XAUUSD").ohlcv(interval="1d", length=5)
display(df_gold)

print("-- 3. Chỉ số Chứng khoán Dow Jones US30 (Interval 4h, 5 nến) --")
# Hỗ trợ mã ngắn USA30, thư viện sẽ tự động phân giải thành USA30.IDX-USD
df_dowjones = mkt.index("USA30", scope="global").ohlcv(interval="4h", length=5)
display(df_dowjones)

print("-- 4. Đổi Múi Giờ Dữ Liệu Thị Trường Quốc Tế --")
# Dukascopy mặc định trả dữ liệu tính theo chuẩn khung giờ Việt Nam (GMT+7).
# Bạn có thể sử dụng param timezone để đưa dữ liệu về trả chuẩn UTC (hoặc bất kỳ múi giờ nào khác)
df_utc = mkt.forex("EURUSD", timezone="UTC").ohlcv(interval="1d", length=5)
display(df_utc)
````
