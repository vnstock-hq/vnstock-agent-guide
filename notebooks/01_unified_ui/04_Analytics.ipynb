````python
from vnstock_data import Analytics
from IPython.display import display

# Initialize Analytics layer
ana = Analytics()
````

### 1. Market Valuation (Định Giá Thị Trường & Chỉ Số)

Truy xuất lịch sử định giá của các rổ chỉ số vĩ mô như VNINDEX, HNX qua các tỷ số kinh điển P/E, P/B và đánh giá tổng quan.

**Lịch sử định giá P/E 5 năm gần nhất của VNINDEX (P/E History)**

````python
df = ana.valuation('VNINDEX').pe(duration='5Y')
display(df.tail())
````

**Lịch sử định giá P/B 3 năm của HNX (P/B History)**

````python
df = ana.valuation('HNX').pb(duration='3Y')
display(df.tail())
````

**Bảng thống kê đánh giá tổng hợp mức độ đắt/rẻ của Thị trường (Market Evaluation Summary)**

````python
df = ana.valuation('VNINDEX').evaluation(duration='1Y')
display(df.head())
````
