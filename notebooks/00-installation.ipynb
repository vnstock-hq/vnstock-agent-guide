Các câu lệnh dưới đây giúp khởi động chương trình cài đặt trên hệ thống máy tính Linux (máy chủ, dịch vụ cloud như Google Colab, Github Codespace hoặc máy tính cá nhân) và macOS trong đó bạn có thể tương tác với chương trình Terminal để nhập thông tin yêu cầu. 

- Sau khi chương trình khởi động, bạn Enter để chọn mặc định hoặc chọn 1 lựa chọn phù hợp mục đích sử dụng của mình.
- Nếu copy câu lệnh ra sử dụng trong Terminal tích hợp hoặc Terminal của máy tính thì bỏ dấu `!` trước câu lệnh

### Máy chủ & máy tính Linux

```python
# Tải file về và lưu với tên vnstock-cli-installer.run
!wget https://vnstocks.com/files/vnstock-cli-installer.run
# Phân quyền và chạy cài đặt
!chmod +x vnstock-cli-installer.run
!./vnstock-cli-installer.run
```

### Macbook (macOS)

Trên macOS không có sẵn công cụ `wget` nên thay thế lệnh wget thành curl

```python
# Tải file về và lưu với tên vnstock-cli-installer.run
!curl -o vnstock-cli-installer.run https://vnstocks.com/files/vnstock-cli-installer.run
# Phân quyền và chạy cài đặt
!chmod +x vnstock-cli-installer.run
!./vnstock-cli-installer.run
```

## Cài đặt với môi trường Terminal không tương tác

Hướng dẫn dưới đây giúp cài đặt chương trình với các hệ thống không thể nhập thông tin trực tiếp vào Terminal mà hoàn toàn dựa vào câu lệnh tự động ví dụ Docker, các dịch vụ serverless như Vercel.

- Bỏ dấu `!` ở đầu dòng khi dùng lệnh này ngoài môi trường Notebook hiện tại
- Thay thế `API_KEY_CỦA_BẠN` ở cuối dòng lệnh bằng giá trị API key thực tế từ tài khoản Vnstock, truy cập mục [API Key của bạn](https://vnstocks.com/account)

```python
!wget -q https://vnstocks.com/files/vnstock-cli-installer.run -O installer.run && chmod +x installer.run && echo "2" | ./installer.run --quiet --accept -- --api-key "API_KEY_CỦA_BẠN"
```
