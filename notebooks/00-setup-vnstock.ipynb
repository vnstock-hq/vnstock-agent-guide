Phương pháp này hoạt động trên Linux và Mac. Nếu không có CLI, sử dụng phương pháp pip fallback ở bên dưới.

```python
# Cài đặt vnstock bằng CLI installer
import os
import subprocess
import platform
import sys

print("=" * 60)
print("VNStock - Cài Đặt Thư Viện")
print("=" * 60)

# Kiểm tra nếu vnstock đã cài đặt
try:
    import vnstock
    print(f"✓ vnstock đã được cài đặt (phiên bản: {vnstock.__version__})")
    print("\nBạn có thể bỏ qua bước cài đặt và chuyển sang phần tiếp theo.")
    INSTALLED = True
except ImportError:
    INSTALLED = False
    print("✗ vnstock chưa được cài đặt. Bắt đầu cài đặt...")

if not INSTALLED:
    # Kiểm tra hệ điều hành
    system = platform.system()
    print(f"\nHệ điều hành: {system}")
    
    if system in ['Linux', 'Darwin']:  # Darwin là macOS
        print("\nTải CLI installer...")
        
        # Tải file installer
        download_cmd = 'wget https://vnstocks.com/files/vnstock-cli-installer.run'
        try:
            os.system(download_cmd)
            print("✓ Tải file thành công")
            
            # Phân quyền
            os.system('chmod +x vnstock-cli-installer.run')
            print("✓ Đã phân quyền file")
            
            # Chạy installer
            print("\nChạy CLI installer...")
            os.system('./vnstock-cli-installer.run')
            print("\n✓ Cài đặt CLI hoàn tất!")
            
        except Exception as e:
            print(f"✗ Lỗi khi cài đặt CLI: {e}")
            print("\nSử dụng phương pháp pip fallback...")
    else:
        print(f"\n⚠ CLI installer không khả dụng trên {system}")
        print("Sử dụng phương pháp pip fallback...")
```

## 2. Cài Đặt Pip (Fallback cho môi trường không có CLI)

Phương pháp này hoạt động trên tất cả các môi trường

```python
# Fallback: Cài đặt bằng pip
import subprocess
import sys

try:
    import vnstock
    print(f"✓ vnstock đã được cài đặt (phiên bản: {vnstock.__version__})")
except ImportError:
    print("Cài đặt vnstock bằng pip...")
    try:
        subprocess.check_call([sys.executable, "-m", "pip", "install", "-U", "vnstock"])
        print("\n✓ Cài đặt pip hoàn tất!")
    except subprocess.CalledProcessError as e:
        print(f"✗ Lỗi cài đặt: {e}")
        print("\nVui lòng kiểm tra kết nối internet hoặc liên hệ hỗ trợ.")
```
